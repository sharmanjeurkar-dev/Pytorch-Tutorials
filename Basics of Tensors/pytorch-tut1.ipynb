### Matrix Multiplication in Tensors

```python
#use torch.matmul() for matrix multiplication torch.mm() shortcut
#shape valid of format :(m,n)*(n,m)

tensor_a = torch.tensor([[1,2],
                        [3,4],
                        [5,6]], dtype = torch.float32)
tensor_b = torch.tensor([[1,2],
                        [3,4],
                        [5,6]], dtype = torch.float32)

#torch_a*torch_b --> shape error -> (3,2)*(3,2)
#Correct method
#tensor_a*tensor_b --> is element by element multiplcation and .matmul() is matrix multiplication
torch.matmul(tensor_a,tensor_b.T)
```

```python
# Since the linear layer starts with a random weights matrix, let's make it reproducible (more on this later)
torch.manual_seed(42)
# This uses matrix multiplication
linear = torch.nn.Linear(in_features=2, # in_features = matches inner dimension of input 
                         out_features=6) # out_features = describes outer value 
x = tensor_a
output = linear(x)
print(f"Input shape: {x.shape}\n")
print(f"Output:\n{output}\n\nOutput shape: {output.shape}")
```

```python
a = torch.rand(size = (7,1))
a
```

```python
#to remove unwanted dimensions or in general delete dimension 1 from the dimension of tensors use torch.squeeze()
a_squeezed = torch.squeeze(a)
a_squeezed#the dimension of 1 removed and elements in the tensors are rearranged
```

```python
#Reverse of squeeze() is torch.unsqueeze() i.e. to add dimension of 1 any where in the tensor or at a praticular dimension vale
a_unsqueezed1 = torch.unsqueeze(a,dim=0)
a_unsqueezed2 = torch.unsqueeze(a,dim=2)# here at dim=1 automatic 1 is added and then 2 
a_unsqueezed1,a_unsqueezed2,a_unsqueezed2.shape
```

```python
y = torch.rand(size = (224,225,5))
```

```python
# to rearrange the dimension use permute 
y_permute = y.permute((2,1,0)) # now 224 is the last dim in shape, 225 is the 2nd dim and 5 is the 1st dim in the shape pf y. The elemenst are hence rearranged in the new shape after using permute
y_permute.shape
```
